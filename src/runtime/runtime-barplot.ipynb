```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 4
runtimes = (0.5, 16.3, 42.4, 207.0)
memory = (24, 114, 114, 2215)

# create plot
fig, ax0 = plt.subplots()
ax1 = ax0.twinx() # Create another axes that shares the same x-axis as ax.
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

color = 'tab:red'
rects1 = ax0.bar(index, runtimes, bar_width,
alpha=opacity,
color=color,
label='Frank')
ax0.tick_params(axis='y', labelcolor=color)
ax0.set_ylabel('Runtime (s)', color=color)


color = 'tab:blue'
rects2 = ax1.bar(index + bar_width, memory, bar_width,
alpha=opacity,
color=color,
label='Guido')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylabel('GPU Memory (MB)', color=color)


plt.xticks(index + bar_width / 2, ('L2', 'SSIM', 'Watson-fft', 'Deeploss-vgg'))

plt.tight_layout()
plt.savefig('runtime.png', dpi=300)
plt
```
